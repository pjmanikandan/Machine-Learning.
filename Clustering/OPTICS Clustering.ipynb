```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python

dataset=pd.get_dummies(dataset,drop_first=True)#get dummies is used to catogorical data handeling and changed numeric data by one hot encoding finaly is gives dummies so droup_first=True is delete first dummie colum
dataset

```

```python
dataset.columns
```

```python
x=dataset.iloc[:,[2,3]].values
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import OPTICS
OPTICS=OPTICS(min_samples=5,  metric='minkowski', p=2, metric_params=None, cluster_method='xi', eps=None, xi=0.05, predecessor_correction=True, min_cluster_size=None, algorithm='auto', leaf_size=30, memory=None, n_jobs=None)
y_OP=OPTICS.fit_predict(x)
```

```python
y_OP
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised
```

```python
supervised["cluster_group"]=y_OP
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
import seaborn as sns
```
