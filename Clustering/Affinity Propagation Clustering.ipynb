```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python

dataset=pd.get_dummies(dataset,drop_first=True)#get dummies is used to catogorical data handeling and changed numeric data by one hot encoding finaly is gives dummies so droup_first=True is delete first dummie colum
dataset

```

```python
dataset.columns
```

```python
x=dataset.iloc[:,[2,3]].values
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AffinityPropagation
AffinityPropagation=AffinityPropagation(damping=0.5, max_iter=200, convergence_iter=15, 
                         copy=True, preference=None, affinity='euclidean', verbose=False)
y_AF=AffinityPropagation.fit_predict(x)
```

```python
y_AF
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised
```

```python
supervised["cluster_group"]=y_AF
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
import seaborn as sns
```

```python
face=sns.lmplot(data=supervised,x=supervised.columns[2],y=supervised.columns[3],hue=supervised.columns[5],
               fit_reg=False,legend=True,legend_out=True)
```
