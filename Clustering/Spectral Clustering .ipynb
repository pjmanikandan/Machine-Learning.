```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python

dataset=pd.get_dummies(dataset,drop_first=True)#get dummies is used to catogorical data handeling and changed numeric data by one hot encoding finaly is gives dummies so droup_first=True is delete first dummie colum
dataset

```

```python
dataset.columns
```

```python
x=dataset.iloc[:,[2,3]].values
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import SpectralClustering
SpectralClustering=SpectralClustering(n_clusters=10,  random_state=None, n_init=10, gamma=1.0, affinity='rbf', n_neighbors=10, eigen_tol=0.0, assign_labels='discretize', degree=3, coef0=1, kernel_params=None, n_jobs=None)
y_SC=SpectralClustering.fit_predict(x)
```

```python
y_SC
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised
```

```python
supervised["cluster_group"]=y_SC
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
import seaborn as sns
```
