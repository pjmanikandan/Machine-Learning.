```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python

dataset=pd.get_dummies(dataset,drop_first=True)#get dummies is used to catogorical data handeling and changed numeric data by one hot encoding finaly is gives dummies so droup_first=True is delete first dummie colum
dataset

```

```python
dataset.columns
```

```python
x=dataset.iloc[:,[2,3]].values
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import Birch
Birch=Birch(threshold=0.5, branching_factor=50, n_clusters=5, compute_labels=True, copy=True)
y_BIRCH=Birch.fit_predict(x)
```

```python
y_BIRCH
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised
```

```python
supervised["cluster_group"]=y_BIRCH
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
import seaborn as sns
```

```python
face=sns.lmplot(data=supervised,x=supervised.columns[2],y=supervised.columns[3],hue=supervised.columns[5],
               fit_reg=False,legend=True,legend_out=True)
```
